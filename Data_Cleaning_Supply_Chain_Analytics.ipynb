## Merge tables

- The most straight forward thing to do is to merge the Inventory Table (dfi) and Fulfillment Table (dff) on "Product Name".

```python
# Visual Python: Data Analysis > Bind
dfif = pd.merge(dfi, dff, on=['Product Name'], how='left').reset_index(drop=True)
dfif
```

## Observation from checking unique values in Order n Shipping Table (dfs) and dfif Table before merging

- The number of "Product Name" in dfs table and dfif table does not tally. 
- dfif table contains the information of all products which are important to tackle the inventory management challenges. 
- Therefore, using a RIGHT join will keep all products for analysis in Tableau. But there will be 175 null values in columns belonging to dfs table after the merge.

```python
# Visual Python: Data Analysis > Bind
merge_df = pd.merge(dfs, dfif, on=['Product Name'], how='right').reset_index(drop=True)
merge_df
```

```python
merge_df.nunique()
```

```python
merge_df.info()
```

```python
merge_df.isnull().sum()
```

## Feature engineering and removing redundant columns

- Redundant columns: OrderYearMonth, Year Month
- Add new features: Total Inventory Cost = Warehouse Inventory * Inventory Cost Per Unit

```python
merge_df.columns
```

```python
merge_df.drop([' Order YearMonth ', ' Year Month '], axis=1, inplace=True)
merge_df.columns
```

```python
merge_df['Total Inventory Cost'] = merge_df[' Warehouse Inventory '] * merge_df['Inventory Cost Per Unit']
merge_df.info()
```

```python
# Visual Python: Data Analysis > File
merge_df.to_csv('./data/cleaned_data.csv')
```
